```python
# import numpy as np
import numpy as np
# import matplotlib
#import matplotlib
import matplotlib.pyplot as plt
```

```python
# Run the following code
def initialize_qubit():
    ket0 = np.array([[1], [0]])
    return ket0
```

```python
# test the function
ket0 = initialize_qubit()
# print the result
print(ket0)
```

```python
# code the function
def apply_hadamard(ket):
    hadamard = np.array([[1, 1], [1, -1]]) / np.sqrt(2)
    return np.dot(hadamard, ket)
```

```python
# test the function
ket0 = initialize_qubit()
ket1 = apply_hadamard(ket0)
print(ket1)
```

```python
# Define the function apply_pauli_x as mentioned in the above text
def apply_pauli_x(ket):
    pauli_x = np.array([[0, 1], [1, 0]])
    return np.dot(pauli_x, ket)
```

```python
# test the function apply_pauli_x
ket0 = initialize_qubit()
ket1 = apply_pauli_x(ket0)
print(ket1)
#print(ket0)
```

```python
# code the measurements function
def measure_qubit(ket):
    prob = np.abs(ket)**2
    result = np.random.choice([0, 1], p=prob.ravel())
    return result
```

```python
psi_0 = initialize_qubit()
psi_1 = apply_hadamard(psi_0)
result = measure_qubit(psi_1)
print(result)
```

```python
# This is the code, you can tailor it to your needs
def simulate_qubit(n_runs):
    results = []
    for i in range(n_runs):
        psi_0 = initialize_qubit()
        psi_1 = apply_hadamard(psi_0)
        result = measure_qubit(psi_1)
        results.append(result)
    return results
```

```python
# test the function with respect to different number of runs: 1, 10, 100, 1000
n_runs = [1, 10, 100, 1000]
for n in n_runs:
    results = simulate_qubit(n)
    print(results)
```

```python
# Import qiskit libraries
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit.circuit import Parameter
```

```python
# Create a quantum register of 1 qubit
qreg = QuantumRegister(1)
```

```python
# create a classical register of 1 bit
creg = ClassicalRegister(1)
```

```python
# Create a quantum circuit with qreg and creg
circuit = QuantumCircuit(qreg, creg)
```

```python
# print the circuit
print(circuit)
```

```python
# Draw the circuit
circuit.draw(output='mpl')
```

```python
# Apply a Hadamard gate on qubit 0
circuit.h(qreg[0])

# Draw the circuit
circuit.draw(output='mpl')
```

```python
# Measure qubit 0
circuit.measure(qreg[0], creg[0])

# print the result
print(circuit)

# Draw the circuit
circuit.draw(output='mpl')
```

```python
 #Get the number of qubits
print(circuit.num_qubits)

# Get the number of classical bits
print(circuit.num_clbits)

# Get the number of gates
print(circuit.size())
```

```python
# First, import the necessary libraries
from qiskit import QuantumCircuit, transpile
from qiskit.providers.basic_provider import BasicSimulator
from qiskit.visualization import plot_histogram
```

```python
# Try to run the circuit
# Create a quantum circuit of 1 qubit and 1 bit
qreg = QuantumRegister(1)
creg = ClassicalRegister(1)
circuit = QuantumCircuit(qreg, creg)

# apply Hadamard gate
circuit.h(qreg[0])

# measure the qubit
circuit.measure(qreg[0], creg[0])

# Specify the backend (the simulator)
simulator = BasicSimulator()

# Transpile the circuit for the simulator
compiled_circuit = transpile(circuit, simulator)

# Run the compiled circuit
job = simulator.run(compiled_circuit)

# Get the result
result = job.result()

# Print the counts
print(result.get_counts())

# Plot the histogram
plot_histogram(result.get_counts())
```
